# ข้อมูลที่สนใจ: ข้อมูล Log การเข้า-ออกหอสมุด (Library Gate Check-in/Out), ข้อมูลการจองห้องอ่านหนังสือ หรือข้อมูลการเชื่อมต่อ Wi-Fi ในหอสมุด เพื่อดูเวลาเริ่มต้นและเวลาสิ้นสุด

2. คำอธิบาย

3. ประโยชน์
 นักศึกษา: รู้ล่วงหน้าว่าโซนไหนคนแน่น ช่วงเวลาไหนที่นั่งเต็ม จะได้วางแผนมาติวกับเพื่อนได้ถูกเวลา ไม่ต้องเดินหาโต๊ะว่างให้เสียเวลา
